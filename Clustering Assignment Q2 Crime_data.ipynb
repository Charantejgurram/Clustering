# Hierachichal Clustering 

```python
#Normalization

def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x) 
```

```python
df_norm = norm_func(df.iloc[:,1:])
df_norm
```

```python
#dendogram using method="single"
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # to create dendrogram  
z = linkage(df_norm, method="single",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
#it is very difficult to form clusters by using these dendogram 
#so, we try  another methods to get better results
```

```python
#dendogram using method="complete"

from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # to create dendrogram  
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
#it is not suggestable to take these method
#so, we go for average method
```

```python
#dendogram using method="average"

from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # to create dendrogram  
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=6, linkage='average',affinity = "euclidean").fit(df_norm) 
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
df['clust']=cluster_labels 
df
```

# Kmeans clustering 

```python
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
```

```python
kc = pd.read_csv('crime_data.csv')
kc
```

```python
#Normalization
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x) 
```

```python
kc_norm = norm_func(kc.iloc[:,1:])  
kc_norm.head(20)
```

```python
fig = plt.figure(figsize=(10, 8))
WCSS = []    
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(kc_norm)
    WCSS.append(clf.inertia_) # inertia another name for wcss
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
#from the  above curve we can find that there is 5 clusters
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(kc_norm)  
```
